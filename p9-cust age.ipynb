```python
import matplotlib.pyplot as plt
import numpy as np

# Sample Data: Customer age and average monthly spending ($)
ages = np.array([22, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 75])
spending = np.array([200, 450, 300, 700, 1200, 600, 400, 1500, 900, 400, 300, 200])

# Define spending categories (low < 400, medium 400-900, high > 900)
categories = []
for spend in spending:
    if spend < 400:
        categories.append('Low')
    elif spend <= 900:
        categories.append('Medium')
    else:
        categories.append('High')

# Map categories to colors and sizes
color_map = {'Low': 'skyblue', 'Medium': 'orange', 'High': 'red'}
size_map = {'Low': 50, 'Medium': 100, 'High': 200}
colors = [color_map[cat] for cat in categories]
sizes = [size_map[cat] for cat in categories]

# ------------------------
# Scatter Plot: Age vs Spending
plt.figure(figsize=(12,5))

plt.subplot(1, 2, 1)
scatter = plt.scatter(ages, spending, c=colors, s=sizes, alpha=0.7, edgecolors='w', linewidth=0.7)

# Label important data points (high spenders)
for i, (age, spend, cat) in enumerate(zip(ages, spending, categories)):
    if cat == 'High':
        plt.text(age, spend + 50, f'Age {age}', fontsize=9, ha='center')

plt.title('Customer Age vs Monthly Spending')
plt.xlabel('Age (years)')
plt.ylabel('Average Monthly Spending ($)')
plt.grid(True, linestyle='--', alpha=0.5)

# Legend manually created
from matplotlib.lines import Line2D
legend_elements = [
    Line2D([0], [0], marker='o', color='w', label='Low Spending', markerfacecolor='skyblue', markersize=8),
    Line2D([0], [0], marker='o', color='w', label='Medium Spending', markerfacecolor='orange', markersize=10),
    Line2D([0], [0], marker='o', color='w', label='High Spending', markerfacecolor='red', markersize=12)
]
plt.legend(handles=legend_elements, title='Spending Category')

# ------------------------
# Pie Chart: Spending Category Distribution
plt.subplot(1, 2, 2)
category_counts = [categories.count(cat) for cat in ['Low', 'Medium', 'High']]
pie_colors = ['skyblue', 'orange', 'red']

# Explode most frequent category
max_index = np.argmax(category_counts)
explode = [0.1 if i == max_index else 0 for i in range(len(category_counts))]

plt.pie(category_counts, labels=['Low', 'Medium', 'High'], colors=pie_colors, autopct='%1.1f%%',
        startangle=90, explode=explode, shadow=True, textprops={'fontsize': 12})

plt.title('Customer Spending Category Distribution')

plt.tight_layout()
plt.show()
```
