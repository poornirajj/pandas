```python
import matplotlib.pyplot as plt
import numpy as np

# Sample Data
departments = ['HR', 'Sales', 'Engineering', 'Marketing', 'Finance', 'IT']
avg_salaries = [55000, 62000, 80000, 58000, 75000, 70000]  # Average salaries per department in $
# Simulated salaries for 300 employees (random normal distribution around average)
np.random.seed(42)
all_salaries = np.concatenate([
    np.random.normal(55000, 5000, 50),  # HR
    np.random.normal(62000, 7000, 60),  # Sales
    np.random.normal(80000, 10000, 80), # Engineering
    np.random.normal(58000, 6000, 40),  # Marketing
    np.random.normal(75000, 8000, 40),  # Finance
    np.random.normal(70000, 7000, 30)   # IT
])
all_salaries = np.clip(all_salaries, 30000, 120000)  # Clip salaries to realistic range

# --- Bar Chart: Average Salary by Department ---
plt.figure(figsize=(10,6))
bars = plt.bar(departments, avg_salaries, color='cornflowerblue', edgecolor='black', width=0.6)

plt.title('Average Salary by Department', fontsize=16)
plt.xlabel('Department', fontsize=12)
plt.ylabel('Average Salary ($)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add value labels on bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1000, f'${int(yval):,}', ha='center', fontsize=10)

plt.ylim(0, max(avg_salaries) + 20000)
plt.tight_layout()
plt.savefig('avg_salary_department.png')
plt.show()

# --- Histogram: Distribution of All Employee Salaries ---
plt.figure(figsize=(10,6))
plt.hist(all_salaries, bins=15, color='mediumseagreen', edgecolor='black', alpha=0.8)

plt.title('Salary Distribution Across All Employees', fontsize=16)
plt.xlabel('Salary Range ($)', fontsize=12)
plt.ylabel('Number of Employees', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xlim(30000, 120000)

plt.tight_layout()
plt.savefig('salary_distribution.png')
plt.show()
```
