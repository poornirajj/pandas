```python
# ✅ Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np

# ✅ Generate sample data: 100 student scores (0–100 scale)
np.random.seed(42)  # For reproducibility
scores = np.random.normal(loc=70, scale=10, size=100)  # Mean 70, Std 10
scores = np.clip(scores, 0, 100)  # Ensure all scores stay within 0–100

# ✅ Create the histogram
plt.figure(figsize=(10, 6))
plt.hist(scores, bins=10, color='skyblue', edgecolor='black', alpha=0.8)

# ✅ Add title and labels
plt.title('Distribution of Math Exam Scores (100 Students)', fontsize=16)
plt.xlabel('Score Range', fontsize=12)
plt.ylabel('Number of Students', fontsize=12)

# ✅ Add grid for better readability
plt.grid(axis='y', linestyle='--', alpha=0.6)

# ✅ Set X and Y axis limits
plt.xlim(0, 100)
plt.ylim(0, 25)  # Adjust based on data

# ✅ Display the plot
plt.tight_layout()
plt.show()
```
