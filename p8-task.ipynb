```python
import matplotlib.pyplot as plt
import numpy as np

# ------------------------------
# Task 2: Sample Sales Data for Scatter Plot
# ------------------------------
products = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
sales_revenue = np.array([50000, 80000, 60000, 40000, 90000])  # in $
profit_margin = np.array([10, 25, 15, 8, 30])  # in %
units_sold = np.array([1500, 3000, 2000, 1000, 3500])

# ------------------------------
# Task 4: Define colors based on profit margin (high profit >= 20)
colors = ['green' if pm >= 20 else 'red' for pm in profit_margin]
sizes = units_sold / 10  # Scale down units sold for marker size

# ------------------------------
# Task 7: Sales Data for Pie Chart
regions = ['North', 'South', 'East', 'West']
regional_sales = [250000, 400000, 150000, 200000]

# Explode the region with highest sales
explode = [0.1 if sale == max(regional_sales) else 0 for sale in regional_sales]

# Colors for pie chart regions
pie_colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']

# ------------------------------
# Task 13: Create Subplots to show both charts together
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# ------------------------------
# Task 3 & 4: Scatter Plot (Sales Revenue vs Profit Margin)
scatter = ax1.scatter(sales_revenue, profit_margin, s=sizes, c=colors, alpha=0.7, edgecolors='w', linewidth=0.5)

# Task 5: Label each scatter point with product name
for i, product in enumerate(products):
    ax1.text(sales_revenue[i], profit_margin[i] + 0.8, product, fontsize=9, ha='center')

# Task 6: Titles, labels, grid
ax1.set_title('Sales Revenue vs Profit Margin', fontsize=14)
ax1.set_xlabel('Sales Revenue ($)')
ax1.set_ylabel('Profit Margin (%)')
ax1.grid(True, linestyle='--', alpha=0.5)

# ------------------------------
# Task 8, 9, 10, 11, 12: Pie Chart of Sales by Region
ax2.pie(regional_sales, labels=regions, colors=pie_colors, autopct='%1.1f%%', startangle=140,
        explode=explode, shadow=True, textprops={'fontsize': 12})

ax2.set_title('Sales Distribution by Region', fontsize=14)

# ------------------------------
# Show the combined figure
plt.tight_layout()
plt.show()
```
