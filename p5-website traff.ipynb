```python
import matplotlib.pyplot as plt
import numpy as np

# Sample Data
pages = ['Home', 'About', 'Services', 'Blog', 'Contact']
daily_visitors = [1500, 600, 1200, 800, 400]  # Visitors per day per page

# Simulate session durations for 500 sessions (in minutes)
np.random.seed(42)
session_durations = np.random.exponential(scale=5, size=500)  # Exponential distribution to simulate session times
session_durations = np.clip(session_durations, 0, 30)  # Limit session times to 30 minutes max

# --- Bar Chart: Daily Visitors per Page ---
plt.figure(figsize=(10,6))
bars = plt.bar(pages, daily_visitors, color='mediumslateblue', edgecolor='black', width=0.6)

plt.title('Daily Visitors per Website Page', fontsize=16)
plt.xlabel('Page', fontsize=12)
plt.ylabel('Number of Visitors', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add visitor numbers on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 30, f'{height}', ha='center', fontsize=10)

plt.ylim(0, max(daily_visitors) + 200)
plt.tight_layout()
plt.savefig('daily_visitors_bar_chart.png')
plt.show()

# --- Histogram: Distribution of Session Durations ---
plt.figure(figsize=(10,6))
plt.hist(session_durations, bins=15, color='darkorange', edgecolor='black', alpha=0.8)

plt.title('Distribution of Session Durations', fontsize=16)
plt.xlabel('Session Duration (minutes)', fontsize=12)
plt.ylabel('Number of Sessions', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xlim(0, 30)

plt.tight_layout()
plt.savefig('session_duration_histogram.png')
plt.show()
```
