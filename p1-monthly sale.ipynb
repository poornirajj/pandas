```python
# ✅ Import necessary library
import matplotlib.pyplot as plt

# ✅ Define the months and corresponding sales data (in $1000s)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

sales = [50, 55, 48, 60, 65, 70, 68, 75, 72, 78, 85, 90]  # Example sales in thousands

# ✅ Create the bar chart
plt.figure(figsize=(12, 6))
bars = plt.bar(months, sales, color='skyblue', edgecolor='black')

# ✅ Add labels and title
plt.title('Monthly Sales Performance - 2024', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Sales Revenue (in $1000s)', fontsize=12)

# ✅ Add grid for better readability
plt.grid(axis='y', linestyle='--', color='gray', alpha=0.5)

# ✅ Annotate bars with values
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, f'{yval}', ha='center', fontsize=10)

# ✅ Set Y-axis limit for better spacing
plt.ylim(0, max(sales) + 15)

# ✅ Show the plot
plt.tight_layout()
plt.show()
```
