```python
import matplotlib.pyplot as plt

# Sample data: Car brands and their market share percentages
brands = ['Toyota', 'Ford', 'Volkswagen', 'Honda', 'BMW', 'Tesla']
market_share = [25, 20, 15, 12, 10, 18]  # Percentages that sum to 100

# Colors for the pie chart
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0','#ffb3e6']

# Explode the largest slice to highlight it
explode = [0.1 if share == max(market_share) else 0 for share in market_share]

# Create pie chart
plt.figure(figsize=(8,8))
plt.pie(market_share, labels=brands, colors=colors, autopct='%1.1f%%',
        startangle=140, explode=explode, shadow=True, textprops={'fontsize': 12})

plt.title('Market Share of Car Brands', fontsize=16)
plt.axis('equal')  # Equal aspect ratio ensures pie is circular

plt.tight_layout()
plt.show()
```
